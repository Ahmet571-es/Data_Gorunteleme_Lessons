```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

#### scatterplot :

```python
plt.scatter(df["total_bill"], df["tip"])
plt.show()
```

```python
# 1 satır 2 sutunluk scatter yapalım : 
fig, ((axes1,axes2)) = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
axes1.scatter(df["tip"], df["total_bill"])
axes2.scatter(df["size"], df["total_bill"])
plt.tight_layout(pad= 5, h_pad= 4)
plt.show()
```

#### barchar :

```python
plt.bar(df["sex"], df["total_bill"])
plt.show()
```

```python
fig, ((axes1,axes2)) = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
axes1.bar(df["sex"], df["total_bill"])
axes2.bar(df["smoker"], df["total_bill"])
plt.tight_layout(pad=5, h_pad=5)
plt.show()
```

```python
# 1 satır 3 sutun cinsiyet, sigara içme, time :
fig, ((axes1),(axes2),(axes3)) = plt.subplots(nrows=1, ncols=3, figsize=(8,4))
axes1.bar(df["sex"], df["total_bill"])
axes2.bar(df["smoker"], df["total_bill"])
axes3.bar(df["time"], df["total_bill"])
plt.tight_layout(pad=5, h_pad=5)
plt.show()
```

```python
# günlere göre para kazanma durumuna bakalım.
plt.bar(df["day"], df["total_bill"])
plt.show()
```

#### pie charts :

```python
# yüzdelik grafiktir. yüzdelik bilgisi verir.
```

```python
a = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(a, labels= mylabels, autopct="%.1f") # %.1f demek % den sonra bir değer al demektir.
plt.show()
```

```python
# cinsiyete göre hesap ödeme durumu
m,f = round(df.groupby("sex")["total_bill"].sum(),2)
hesap = np.array([m,f])
cinsiyet = ["Male","Female"]
my_explode =[0.1,0]
plt.pie(hesap, labels=cinsiyet, autopct="%1.f", explode =my_explode)
plt.show()
```

```python
# cinsiyete göre bahşiş verme : 
m,f = round(df.groupby("sex")["tip"].sum())
bahsis = np.array([m,f])
my_label = ["Male","Female"]
my_explode =[0.1,0]
plt.pie(bahsis, labels= my_label, explode = my_explode, shadow=True, autopct="%.1f")
plt.show()
```

```python
# 1 satır 2 sutun :
m_total, f_total = df.groupby("sex")["total_bill"].sum()
m_tip, f_tip = df.groupby("sex")["tip"].sum()
hesap = np.array([m_total, f_total])
bahsis = np.array([m_tip, f_tip])
my_labels = ["Male","Female"]
my_explode =[0.1,0]
fig, ((axes1,axes2)) = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
axes1.pie(hesap, labels=my_labels, autopct="%.1f", explode =my_explode)
axes2.pie(bahsis, labels=my_labels, autopct="%.1f", explode =my_explode)
plt.tight_layout(pad=5, h_pad=4)
plt.show()
```
