```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

#### countplot :

```python
sns.countplot(x="day", data=tips);
# günlerin count değerlerini verir.
```

```python
tips.day.value_counts()
# yukarıdaki grafiğin kodu
```

```python
sns.countplot(x="day", hue="sex", data=tips);
# günlerin cinsiyete göre count değerlerini verir.
```

```python
tips.groupby("day")["sex"].value_counts().unstack()
# yukarıdaki grafiğin kodu.
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(x="day", data=tips)
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+0.8));
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(x = "day", hue="sex", data=tips)
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x() + 0.1, p.get_height() + 0.6)); # kaç atne sayı olarak ekledik.
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(x = "day", hue="sex", data=tips)
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x() + 0.1, p.get_height() + 0.6));    
    ax.text(p.get_x()+0.1, p.get_height()-3,str(round(p.get_height()/tips.day.count(),2)) ) # yüzdelik değer ekledik
```

```python
mpg=sns.load_dataset('mpg')
mpg.head()
```

```python
fig, ax = plt.subplots()
sns.countplot(x="origin", data=mpg)
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x() + 0.1, p.get_height() + 0.6));
```

```python
fig, ax = plt.subplots()
sns.countplot(x="model_year", hue="origin", data=mpg)
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x() - 0.1, p.get_height() + 0.2));
```

#### Boxplot : 

```python
sns.boxplot(x="time", y="total_bill", data=tips);
```

```python
fig, ax = plt.subplots(figsize=(7,5))
ax = sns.boxplot(x="sex", y="total_bill", hue="time", data=tips);
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) ;
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax = sns.boxplot(x="sex", y="total_bill", hue="day", data=tips, width=0.9); # width kutunun genişliği.
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) ;
```

#### lineplot :

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(x="month", y="passengers", data=flights);
```

```python
f_sum = flights.groupby(["year","month"])[["passengers"]].sum()
f_sum
```

```python
sns.lineplot(x= f_sum.reset_index().index, y = f_sum.passengers);
# yukarıda yaptığımız groupbyın year kısmı indexti onu resetledik ve gösterdik.
```

```python
sns.lineplot(x="year", y="passengers", data = flights[flights["month"] == "May"]);
```

```python
flights_wide = flights.pivot(index = "year", columns = "month", values = "passengers")
flights_wide
# burda yıllara ve aylara göre toplam uçuş sayısını tablo olarak yaptık.
```

```python
sns.lineplot(data = flights_wide);
```
