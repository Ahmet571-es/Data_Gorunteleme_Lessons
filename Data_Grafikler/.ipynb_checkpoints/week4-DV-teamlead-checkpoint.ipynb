```python
import matplotlib.pyplot as plt 
import numpy as np

labels = ['UK', 'USA', 'GER', 'JAP', 'FRA']
age_men = [40, 35, 45, 53, 28]
age_women = [35, 32, 48, 48, 31]
```

```python
#1. soru BENİM ÇÖZÜM : 
fig, ax = plt.subplots()
ax.set_ylabel('Ages')
ax.set_title('Age means by Gender')
ax.bar(labels, age_men, label="Men")
plt.legend()
for i in range(len(labels)):
    plt.text(labels[i], age_men[i], str(age_men[i]), ha="center",va="bottom")
```

```python
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x, age_men, width, label='Men')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Ages')
ax.set_title('Age means by Gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)


fig.tight_layout()

plt.show()
```

```python
# 2. soru BENİM ÇÖZÜM : 
fig, ax = plt.subplots()
ax.set_ylabel('Ages')
ax.set_title('Bayanların Yaşları')
ax.bar(labels, age_women, label="Bayan")
plt.legend()
for i in range(len(labels)):
    plt.text(labels[i], age_women[i], str(age_women[i]), ha="center",va="bottom")
```
