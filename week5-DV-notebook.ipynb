```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
print(sns.get_dataset_names())
```

```python
mpg = sns.load_dataset("mpg")
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
#describe all features, both numerical and categorical
# hem sayısal hem de categoric tüm özellikleri tanımlayın.
titanic.describe(include="all")
```

```python
# get number of null values per columns
# her sutunda toplamda kaç tane null değer var 
titanic.isnull().sum()
```

```python
# we can see all missing value in heatmap
# heatmap ile eksik değerleri görmek istersek 
sns.heatmap(titanic.isnull());
```

```python
# we can see numeric column's corr in heatmap
# titanic veri setinde ki verilerin birbirleri ile olan korelasyon katsayı değerleri.
sns.heatmap(titanic.corr(), annot=True);
```
